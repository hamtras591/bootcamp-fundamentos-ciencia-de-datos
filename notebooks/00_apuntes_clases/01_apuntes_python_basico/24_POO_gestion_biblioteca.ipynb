# Ejercicio 1. Libreria

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.available = True

    def borrow(self):
        if self.available:
            self.available = False
            print(f"El libro {self.title} ha sido alquilado")
        else:
            print(f"El libro {self.title} no esta disponible")

    def return_book(self):
        self.available = True
        print(f"El libro {self.title} ha sido devuelto")

class User:
    def __init__(self, name, user_id):
        self.name = name
        self.user_id = user_id
        self.borrowed_books = []

    def borrow_book(self, book):
        if book.available:
            book.borrow()
            self.borrowed_books.append(book)
        else:
            print(f"El libro {book.title} no esta disponible")

    def return_book(self, book):
        if book in self.borrowed_books:
            book.return_book()
            self.borrowed_books.remove(book)
        else:
            print(f"El libro {book.title} no esta disponible")

class Library:
    def __init__(self):
        self.books = []
        self.users = []

    def add_book(self, book):
        self.books.append(book)
        print(f"El libro {book.title} ha sido agregado a la biblioteca")

    def register_user(self, user):
        self.users.append(user)
        print(f"El usuario {user.name} ha sido registrado")

    def show_available_books(self):
        print(f"Libros dispobibles")
        for book in self.books:
            if book.available:
                print(f"El libro {book.title} esta disponible")

# Creación de propiedades "LIBROS"
book1 = Book("El principito", 'Antoine de Saint-Exupery')
book2 = Book("1984","George Orwell")

print("---")

# Creación de propiedades "USUARIOS"

user1 = User("Anderson Rubio", "001")

print("---")

# Creación Propiedades "LIBRERIA"

Library = Library()
Library.add_book(book1)
Library.add_book(book2)

print("---")
# Mostrar libros disponibles
Library.show_available_books()

print("---")
# Realizar un prestamo
user1.borrow_book(book1)

print("---")
# Mostrar libros disponibles
Library.show_available_books()
print("---")

# Realizar un devolucion
user1.return_book(book1)

print("---")
# Mostrar libros disponibles
Library.show_available_books()

print("---")
```
